```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
#reshape and view
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
#Reshape if we don't know the shape of the items using -1
my_torch2 =torch.arange(15)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(3,-1)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.view(2,5)
my_torch4
```

```python
#with reshape and view, they will update.

my_torch4[1] = 7
my_torch4
```

```python
my_torch3[0] = 1
my_torch3
```

```python
#slices
my_torch5 = torch.arange(10)
my_torch5
```

```python
#grab a specific item
my_torch5[8]
```

```python
#grab slice
my_torch6 = my_torch5.reshape(5,2)
my_torch6
```

```python
my_torch6[:, 1]
```

```python
#return column
my_torch6[:,1:]
```
