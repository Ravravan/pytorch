```python
import torch
```

#TENSOR MATH OPERASTIONS

```python
#Addition
tensor_a = torch.tensor([1,2,3,4,5,6,9])
tensor_b = torch.tensor([4,5,8,2,1,7,3])


tensor_a + tensor_b
```

```python
#Addition longhand
torch.add(tensor_a, tensor_b)
```

```python
#Substrsction
tensor_b - tensor_a
```

```python
#Sub Function
torch.sub(tensor_a, tensor_b)
```

```python
#Multiplication
tensor_a * tensor_b
```

```python
#Mul longhand
torch.mul(tensor_a, tensor_b)
```

```python
#Division
tensor_b / tensor_a
```

```python
#Div longhand
torch.div(tensor_a, tensor_b)
```

```python
#Remainder Moduls
tensor_b % tensor_a
```

```python
#Remainder longhand
torch.remainder(tensor_b, tensor_a)
```

```python
#Exponents
tensor_a ** tensor_b
```

```python
#Exponents longhand
torch.pow(tensor_b, tensor_a)
```

```python
#Reassigment_
tensor_a.add_(tensor_b)
```
